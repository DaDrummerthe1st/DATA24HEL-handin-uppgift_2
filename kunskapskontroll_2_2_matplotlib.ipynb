### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

Figure - The canvas on which each Axes (the actual graph) is drawn using axis (ie x,y,z - width, height, depth in a 3 dimensonal Axes). Artists are all the visual elements (included but not limited to axis, legends, Axes, lines etc)

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# pyplot approach
x = np.linspace(-4, 4, 10)
y = x**2

#plt.figure(figsize=(10,10))
plt.plot(y)

plt.ylabel('y')
plt.xlabel('x')
plt.suptitle('$y = x^2$')
```

```python
# objektorienterad approach
fig, ax = plt.subplots()

ax.plot(y)
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
# create a 1 by 2 figure combining the two different diagrams
fig, axs = plt.subplots(1,3,figsize=(21,7))

fig.suptitle('Fruit sales KPI', fontsize=24)

# first figure shows the scattered diagram
axs[0].scatter(random_data_x, random_data_y)
axs[0].title.set_text('Support success evaluation')
axs[0].set_ylabel('customer complaints compared to last year avg')
axs[0].set_xlabel('customer support grade compared to last year avg')

# second figure is a linear diagram
axs[1].plot(x,y)
axs[1].title.set_text('Waiting customers (incall)')
axs[1].set_ylabel('Number of customers waiting')
axs[1].set_xlabel('minutes waiting compared to last year')

# third figure is a bar diagramlast year
axs[2].bar(names, values)
axs[2].title.set_text('Avg amount (kg) sold per day')
axs[2].get_children()[0].set_color('black')
colors = ['purple','green','orange','yellow','lightgreen']
for child, color, amount in zip(
        axs[2].get_children(), 
        colors,
        values
    ):
    child.set_color(color)
    axs[2].text(# Since everything is plotted (positioned in a specific x, y - place) the text positions has to be calculated
                    (child.get_x()+child.get_width()/4), # x - axis (height)
                    (child.get_y()+child.get_height()/2), # y - axis (width)
                amount,
                color='black',
                size='xx-large',
)
```
